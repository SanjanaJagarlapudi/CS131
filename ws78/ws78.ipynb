```python
from pyspark.sql import SparkSession
import pandas as pd
import matplotlib.pyplot as plt
import time



# Create a SparkSession instance (an entry point to all Spark functions)
spark = SparkSession.builder.appName("w8").getOrCreate()

results = []

#First datafile
file_size_bytes = 18
start = time.time()
# Read a file in CSV format into Spark DataFrame

df = spark.read.csv("gs://dataproc-staging-us-central1-721720945833-6cwqbnms/Datafile1 - Sheet1.csv", header=True, inferSchema=True)

# Groups the DataFrame using the specified columns and returns the average of the
df.groupBy('value1').avg('value2').show()

end = time.time()
total = end - start

results.append(("Datafile1 - Sheet1.csv", file_size_bytes, total))


#Second datafile
file_size_bytes = 38
start = time.time()
# Read a file in CSV format into Spark DataFrame
df = spark.read.csv("gs://dataproc-staging-us-central1-721720945833-6cwqbnms/Datafile2 - Sheet1.csv", header=True, inferSchema=True)

# Groups the DataFrame using the specified columns and returns the average of the
df.groupBy('value1').avg('value2').show()

end = time.time()
total = end - start

results.append(("Datafile2 - Sheet1.csv", file_size_bytes, total))


#Third datafile
file_size_bytes = 63
start = time.time()
# Read a file in CSV format into Spark DataFrame
df = spark.read.csv("gs://dataproc-staging-us-central1-721720945833-6cwqbnms/Datafile3 - Sheet1.csv", header=True, inferSchema=True)

# Groups the DataFrame using the specified columns and returns the average of the
df.groupBy('value1').avg('value2').show()

end = time.time()
total = end - start

results.append(("Datafile3 - Sheet1.csv", file_size_bytes, total))


#Fourth datafile
file_size_bytes = 113
start = time.time()
# Read a file in CSV format into Spark DataFrame
df = spark.read.csv("gs://dataproc-staging-us-central1-721720945833-6cwqbnms/Datafile4 - Sheet1.csv", header=True, inferSchema=True)

# Groups the DataFrame using the specified columns and returns the average of the
df.groupBy('value1').avg('value2').show()

end = time.time()
total = end - start

results.append(("Datafile4 - Sheet1.csv", file_size_bytes, total))


#Fifth datafile
file_size_bytes = 218
start = time.time()
# Read a file in CSV format into Spark DataFrame
df = spark.read.csv("gs://dataproc-staging-us-central1-721720945833-6cwqbnms/Datafile5 - Sheet1.csv", header=True, inferSchema=True)

# Groups the DataFrame using the specified columns and returns the average of the
df.groupBy('value1').avg('value2').show()

end = time.time()
total = end - start

results.append(("Datafile5 - Sheet1.csv", file_size_bytes, total))


results_df = pd.DataFrame(results, columns=["File", "Size_Bytes", "Runtime_in_sec"])


plt.figure(figsize=(11,7))
plt.plot(results_df["Size_Bytes"], results_df["Runtime_in_sec"], marker='*')
plt.title("Runtime of Spark vs File Size")
plt.xlabel("File Size in Bytes")
plt.ylabel("Runtime in seconds")
plt.grid(True)
plt.savefig("runtime_plot.png")
plt.show()
```
